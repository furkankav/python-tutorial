```python
from geopy.geocoders import ArcGIS
nom = ArcGIS()
home = nom.geocode("6665 Santa Cristina, The Colony, TX 75056")
home
# home.latitude # => 33.05838963738087
# home.longitude # => -96.84449824463718
# type(home) # => geopy.location.Location
```

```python
# In order to use geocode in our table we need to manipulate data
df1["Address"] = df1["Address"] + ", " + df1["City"] + ", " + df1["State"] + ", " + df1["Country"]
df1
```

```python
df1["Coordinates"] = df1["Address"].apply(nom.geocode)
df1
```

```python
#df1.Coordinates[0]
df1.Coordinates[3].latitude
# Lambda will be used since we want to use method of coordinates
df1["Latitude"] = df1["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df1["Longitude"] = df1["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df1
```
